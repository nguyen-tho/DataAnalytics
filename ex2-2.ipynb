```python
import findspark
import pyspark

from pyspark.sql import SparkSession
findspark.init()

spark = SparkSession.builder.appName('youtube_trending').getOrCreate()
```

```python
#read csv
path = 'dataset\youtube_trending'
#read CAvideo.csv file
data = spark.read.csv(path+'\CAvideos.csv', inferSchema=True, header=True)
```

```python
data.limit(5).toPandas()
```

```python
data.printSchema()
```

```python
import pyspark.sql.types as T
from pyspark.sql.functions import col, cast, to_date

data = data.withColumn("views", col("views").cast(T.IntegerType()))\
    .withColumn("likes", col("likes").cast(T.IntegerType()))\
    .withColumn("trending_date",to_date(col("trending_date"), 'yy.dd.MM'))\
    .withColumn("dislikes", col("dislikes").cast(T.IntegerType()))\
    .withColumn("comment_count", col("comment_count").cast(T.IntegerType()))\
    .withColumn("comment_disabled", col("comment_disabled").cast(T.BooleanType()))
data.printSchema()
```
